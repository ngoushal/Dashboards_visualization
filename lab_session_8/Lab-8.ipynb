### I Found a website which explains the different domains which you can pursue afer MSIS and the salary ranges of each. So, I thought of extracting that and making a graph out of it for prospective students.

```python
response3 = requests.get("https://www.payscale.com/research/US/Degree=Master_of_Information_Science_(MIS)/Salary")
html3 = response3.content
soup = BeautifulSoup(html3, 'html.parser')
soup.prettify(formatter=lambda s: s.replace(u'\xa0', ' '))
#print (soup.prettify())
pay = soup.find_all("div", {"class":"bar-chart-bootstrap"})
for i in pay:
    print(i.text)
```

```python
for mytable in pay:
    table_body = mytable.find('tbody')
    try:
        rows = table_body.find_all('tr')
        for tr in rows:
            cols = tr.find_all('td')
            for td in cols:
                print(td.text)
    except:
        print("no tbody")
```
